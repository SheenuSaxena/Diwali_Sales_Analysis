# Data Exploratory Analysis

## GENDER

```python
# to show the no. of males and females on bar chart 
gender_counts = sns.countplot(x = 'Gender',data = df)
```

```python
gender_counts = sns.countplot(x = 'Gender',data = df)  # countplot shows the no. of occurence of each categorical variable
for bars in gender_counts.containers:                  # this will iterate over each bar in the countplot
    gender_counts.bar_label(bars)                      # it will gives labels of the total counts to each bar


# NOTE: containers is a graphical object that holds the visual elements of a plot i.e bars in this case 
```

```python
# grouping  the rows on the basis of gender
df.groupby('Gender')
```

```python
#using sum() func. to sum up the amount and orders of the grouped data

df.groupby('Gender',as_index = False)[['Amount', 'Orders']].sum()

#NOTE:if i will not do as_index = False then, python will consider Gender as index which we do not want to do  
```

```python
# sorting the group by data
gender_groupby = df.groupby('Gender',as_index=False)[['Amount', 'Orders']].sum().sort_values(by = ['Amount','Orders'],ascending = False)
```

```python
gender_groupby
```

```python
sns.barplot(x='Gender', y = 'Amount', data = gender_groupby )
```

## Age

```python
Age_count = sns.countplot(x ='Age Group',data = df,hue ='Gender')
# used hue to divide the age group into male and female
```

```python
# to get the data label at each bar
Age_count = sns.countplot(x ='Age Group',data = df,hue ='Gender')
for bars in Age_count.containers:
    Age_count.bar_label(bars)
```

```python
sales_age_group = df.groupby('Age Group',as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
```

```python
sales_age_group
```

```python
sns.barplot(x='Age Group',y='Amount',data=sales_age_group)
```

## States

```python
#total no. of orders from top 10 states
orders_by_state = df.groupby('State',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
orders_by_state
```

```python
sns.barplot(x='State',y='Orders',data=orders_by_state)
```

```python
#to remove the overlapping we will set fisize of the chart

sns.set(rc={'figure.figsize':(16,5)})

#NOTE:
#sns.set(): This is a function allows you to customize the appearance of your plots. 
#It can be used to set various parameters, such as figure size, color palettes, grid styles, etc.
#rc={'figure.figsize': (16, 5)}: This parameter within sns.set() is a dictionary 
```

```python
#to change the color palette
sns.set_palette('bright')  
```

```python
sns.barplot(x='State',y='Orders',data=orders_by_state)
```

```python
#total amount of sales from top 10 states
total_amount_spent = df.groupby('State',as_index=False)['Amount'].sum().sort_values(by='State',ascending=False).head(10)
```

```python
total_amount_spent
```

```python
amt_spent_chart = sns.barplot(data=total_amount_spent,x='State',y='Amount')
```

## Marital Status

```python
married_count = sns.countplot(data=df,x='Marital_Status')
married_count.set_xticklabels(['married', 'unmarried'])
for bars in married_count.containers:
    married_count.bar_label(bars)

sns.set({'figure.figsize':(5,5)})
sns.set_palette('deep')  
#NOTE:Here I labeled the 1 as unmarried and 0 as married
```

```python
# grouped the data on the basis of marital status and then gender

married_status_spent = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
married_status_spent
```

```python
marital_status_chart = sns.barplot(data=married_status_spent,x='Marital_Status',y='Amount',hue='Gender')
marital_status_chart.set_xticklabels(['married','unmarried'])
```
